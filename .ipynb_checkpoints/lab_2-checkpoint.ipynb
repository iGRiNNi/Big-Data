#### Постройте круговую диаграмму количества игр в разрезе жанров. Подпишите каждый сектор. Какой жанр самый популярный?

```python
genre_counts = data['Genre'].value_counts()
plt.pie(genre_counts, labels=data.groupby('Genre').count().index, autopct='%.f%%')
plt.title('Распределение жанров')
```

##### Ответ: <span style="color:red">Action: 20%</span>

```python
genre_percentages = (genre_counts / len(data) * 100).round(1)

for genre, count in genre_counts.items():
    percentage = genre_percentages[genre]
    print(f"{genre}: {count} игр ({percentage}%)")
```

```python
genre_counts.idxmax(), genre_counts.max(), genre_percentages[genre_counts.idxmax()]
```

#### Постройте столбчатую диаграмму количества игр в разрезе игровых платформ. Столбцы должны идти в порядке убывания количества игр. Под какую платформу выпустили больше всего игр?

```python
platform_counts = data['Platform'].value_counts()
plt.figure(figsize=(20, 10))

bars = plt.bar(x=platform_counts.index, height=platform_counts.values)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(int(height)),
            ha='center', va='bottom')

plt.title('Название диаграммы')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

```python
 data['Platform'].value_counts()
```

##### Ответ: <span style="color:red">Под платформу DS</span>

```python
platform_counts.idxmax(), platform_counts.max()
```

#### Постройте столбчатую диаграмму количества игр по годам. Столбцы должны идти в хронологическом порядке. Какая пятилетка была самой удачной для игростроя?

```python
plt.figure(figsize=(20, 10))
years = data['Year'].value_counts().sort_index()
bars = plt.bar(data.groupby('Year').count().index, data.groupby('Year')['Rank'].count())

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(int(height)),
            ha='center', va='bottom')
    
plt.title('Название диаграммы')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

##### Ответ: <span style="color:red">С 2007 по 2011 года</span>

```python
years
```

```python
rolling_5 = years.rolling(window=5).sum()
rolling_5
```

```python
top_group_start = int(rolling_5.idxmax() - 4)
top_group_end = int(rolling_5.idxmax())

top_group_label = f"{top_group_start} - {top_group_end}"
top_value = int(rolling_5.max())

top_group_label, top_value
```

#### Постройте столбчатую диаграмму общих доходов игровой индустрии по годам. Столбцы должны идти в хронологическом порядке. Какой год был самым прибыльным?

```python
year_global_sales = data.groupby('Year')['Global_Sales'].sum()
plt.figure(figsize=(20, 10))
bars = plt.bar(data.groupby('Year').count().index, year_global_sales)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(int(height)),
            ha='center', va='bottom')
    
plt.title('Название диаграммы')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

##### Ответ: <span style="color:red">2008 год</span>

```python
year_global_sales
```

```python
year_global_sales.idxmax(), year_global_sales.max()
```
