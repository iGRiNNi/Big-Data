#### Оцените количество пропусков по каждому из признаков. Для числовых признаков заполните пропуски медианным значением, для прочих признаков удаляйте строки.

```python
print('\nКоличество пропусков до обработки:')
print(data.isnull().sum())
```

```python
##copy_data = data.copy(deep=True)

numeric_cols = data.select_dtypes(include='number').columns
other_cols = data.columns.difference(numeric_cols)

data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].median())
```

```python
data = data.dropna(subset=other_cols)

print('\nКоличество пропусков после обработки:')
print(data.isnull().sum())
#data.shape[0], data.shape[0]
```

#### Постройте круговую диаграмму количества игр в разрезе жанров. Подпишите каждый сектор. Какой жанр самый популярный?

```python
genre_counts = data.groupby('Genre').size()
plt.figure(figsize=(18, 8))

print(genre_counts.idxmax())

plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}, colors=plt.cm.Set3_r.colors)
plt.title('Распределение игр по жанрам')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
genre_counts.idxmax()
```

Ответ: <span style="color:green">Action</span>

#### Постройте столбчатую диаграмму количества игр в разрезе игровых платформ. Столбцы должны идти в порядке убывания количества игр. Под какую платформу выпустили больше всего игр?

```python
platform_counts = data['Platform'].value_counts(sort=True)
plt.figure(figsize=(20, 10))

print(platform_counts.idxmax())

bars = plt.bar(platform_counts.index, platform_counts.values,  color='green')

plt.title('Количество игр по платформам',fontsize=14)
plt.xlabel('Платформа',fontsize=14)
plt.ylabel('Количество игр',fontsize=14)
plt.xticks(rotation=0, fontsize=12)
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        int(height),
        ha='center',
        va='bottom',
        fontsize=12
    )
plt.show()
```

Ответ: <span style="color:green">**PS2**</span>

#### Постройте столбчатую диаграмму количества игр по годам. Столбцы должны идти в хронологическом порядке. Какая пятилетка была самой удачной для игростроя?

```python
year_counts = data['Year'].value_counts().sort_index()
plt.figure(figsize=(20, 10))

bars = plt.bar(year_counts.index, year_counts.values, color='blue')

plt.title('Количество игр по годам',fontsize=14)
plt.xlabel('Год',fontsize=14)
plt.ylabel('Количество игр',fontsize=14)
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        int(height),
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(rotation=0)
plt.show()
```

```python
data['Year'] = data['Year'].astype(int)
min_year = data['Year'].min()

# начало пятилетки от первого года датасета
data['Five_Year_Period'] = min_year + ((data['Year'] - min_year) // 5) * 5

# Количество игр по пятилеткам
five_year_counts = data['Five_Year_Period'].value_counts().sort_index()

# Лучшая пятилетка
best_period_start = five_year_counts.idxmax()
best_period_end = best_period_start + 4
best_count = five_year_counts.max()

print(f'Первая пятилетка начинается с {min_year} года')
print(f'Самая удачная пятилетка: {best_period_start}-{best_period_end}')
print(f'Количество игр: {best_count}')
```

Ответ: <span style="color:green">**2005-2009**</span>

#### Пятилетка через скользящее окно в 5 лет

```python

rolling_5 = year_counts.rolling(window=5).sum()

top_group_start = int(rolling_5.idxmax() - 4)
top_group_end = int(rolling_5.idxmax())

top_group_label = f"{top_group_start} - {top_group_end}"
top_value = int(rolling_5.max())

top_group_label, top_value
```

#### Постройте столбчатую диаграмму общих доходов игровой индустрии по годам. Столбцы должны идти в хронологическом порядке. Какой год был самым прибыльным?

```python
year_sales = data.groupby('Year')['Global_Sales'].sum().sort_index()
plt.figure(figsize=(20, 10))

bars = plt.bar(year_sales.index, year_sales.values, color='blue')

plt.title('Общие мировые продажи игровой индустрии по годам',fontsize=14)
plt.xlabel('Год',fontsize=14)
plt.ylabel('Продажи',fontsize=14)
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        int(height),
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(rotation=0)
plt.show()
```

```python
best_year = year_sales.idxmax()
best_sales = year_sales.max()

print(f'Самый прибыльный год: {best_year}')
print(f'Общие продажи: {best_sales:.2f}')
```

Ответ: <span style="color:green">**2007**</span>

#### Постройте линейный график объемов продаж шутеров в Японии по годам. Какой год был самым прибыльным?

```python
shooters = data[data['Genre'] == 'Shooter'].copy()

shooter_jp_sales = shooters.groupby('Year')['JP_Sales'].sum().sort_index()

best_year = shooter_jp_sales.idxmax()
best_sales = shooter_jp_sales.max()

print(f'Самый прибыльный год для шутеров в Японии: {best_year}')
print(f'Продажи в Японии: {best_sales:.2f}')

plt.plot(shooter_jp_sales.index, shooter_jp_sales.values)
plt.show()
```

Ответ: <span style="color:green">**1995**</span>
